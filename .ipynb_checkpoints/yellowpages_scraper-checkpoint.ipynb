```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
from datetime import datetime
import time

def scrape_yellowpages_rendered(url):
    options = Options()
    options.add_argument("--headless=new")
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")

    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
    driver.get(url)
    time.sleep(5)  # Let the page load

    listings = driver.find_elements(By.CLASS_NAME, "result")

    data = []

    for listing in listings:
        try:
            company = listing.find_element(By.CLASS_NAME, "business-name").text
            phone = listing.find_element(By.CLASS_NAME, "phones").text
            address = listing.find_element(By.CLASS_NAME, "adr").text
            website_link = listing.find_element(By.PARTIAL_LINK_TEXT, "Website").get_attribute("href") if "Website" in listing.text else "N/A"
            detail_link = listing.find_element(By.CLASS_NAME, "business-name").get_attribute("href")

            data.append({
                "Company Name": company,
                "Industry": "Construction",
                "Contact Name": "",
                "Email Address": "",
                "Phone Number": phone,
                "Website URL": website_link,
                "Address": address,
                "Date Added": datetime.now().strftime("%-m/%-d/%y"),
                "Date Contacted": "",
                "Contacted": "",
                "Notes": detail_link,
                "My Offering": "$1,500 setup $20 per month (blog, landing & contact form)"
            })
        except Exception as e:
            print("Skipping due to error:", e)

    driver.quit()
    return pd.DataFrame(data)

url = "https://www.yellowpages.com/brooklyn-ny/general-contractors"
df = scrape_yellowpages_rendered(url)
df.to_excel("yellow_pages_general_contractors.xlsx", index=False)
```
